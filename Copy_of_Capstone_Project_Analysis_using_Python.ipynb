#### **The Society Literacy Development Index**

**Definition**.
The Society Literacy Development Index (*Indeks Pembangunan Literasi Masyarakat/IPLM*) is a
measurement of the efforts carried out by local governments (provincial and regency/municipal levels) in fostering and developing libraries as a vehicle for lifelong learning to achieve a community literacy culture.

The formulation of the Society Literacy Development Index
has been prepared by the Center of Library Development and Reading Interest Correctional (*Pusat Pengembangan Perpustakaan dan Pemasyarakatan Minat Baca/P3MB*) National Library of Indonesia since 2018, which involves two primary aspects, namely:
* Society Aspect (*Aspek Masyarakat/AM*): the total population of the local area according to BPS-Statistics Indonesia data.
* Elements of Society Literacy Development (*Unsur Pembangunan Literasi Masyarakat/UPLM*) consist of seven components: Library Service Equity, Adequacy of Library Collection, Adequacy of Library Staff, Society Visit Level per Day, SNP-based Library, Society Involvement in Socialization Activities, and Library Members.

**Formulation**. The implementation of the Society Literacy Development
Index Study is to determine the condition of all types of
libraries, ranging from aspects of library distribution,
collections, library staff, users, and library members
throughout Indonesia. The Society Literacy Development
Index is calculated by the following formula.

\begin{align}
\mathbf{IPLM} = \sum_{i} \frac{UPLM_i \,}{AM} × 100
\end{align}

Where:
* UPLM: component variable that forms the index of the
society literacy development elements
* AM: total population of the local area

**Composition**. The Society Literacy Development
Index has seven composers as follows.
1. **Library Service Equity**, a comparison between the availability of libraries
nationally (consisting of public libraries, school libraries,
university libraries, and special libraries) with the total
population in each region in Indonesia.
2. **Adequacy of Library Collection**, seen using references from
IFLA/UNESCO, where the ratio of the population in a
country to the ideal number of library collections is 1 : 2.
3. **Adequacy of Library Staff**, the ratio between the
population and the number of library staff, which refers
to the IFLA Standard 2006 public library staff adequacy
ratio, which is 1: 2,500 (standard value <2,500).
4. **Society Visit Level per Day**, the ratio between
the number of visits per day and the total population.
5. **SNP-based Library**, a comparison
between the total number of libraries perceived as having
SNP with the total number of libraries in Indonesia.
6. **Society Involvement in Socialization Activities**, a
comparison between the number of outreach activities
and the total number of libraries.
7. **Library Members**, the ratio between the number of
library members and the total population.

#### **The Level of Reading Interest**

**Definition**.
The Level of Reading Interest (*Tingkat Kegemaran Membaca/TGM*) is the level of
behavior or habits of society in obtaining knowledge
and information from various forms of media that can
be done independently by local governments in a certain
period.

**Formulation**. The Level of Reading Interest is calculated by the following formula.

\begin{align}
\mathbf{TGM} = (0.3 TFM + 0.3 TDM + 0.3 TJB) + (0.05 FAI + 0.05 DAI)
\end{align}

Where:
* TFM: Reading Frequency per week
* TDM: Reading duration per day
* TJB: Number of Reading Material per day
* FAI: Internet Access Frequency per week
* DAI: Internet Access Duration per day

The Level of Reading Interest is categorized as follows.
1. Very low: TGM Value Interval 0–20
2. Low: TGM Value Interval 20.1–40
3. Medium: TGM Value Interval 40.1–60
4. High: TGM Value Interval 60.1–80
5. Very High: TGM Value Interval 80.1–100

**Composition**. The Level of Reading Interest has five composers as follows.
1. **Reading Frequency per week**, the number of reading
activities per week.
2. **Reading duration per day**, the length of time reading
activity per day.
3. **Number of Reading Material per quarterly**, the number
of books completed per three months.
4. **Internet Access Frequency per week**, the frequency of
internet use in general.
5. **Internet Access Duration per day**, the length of time
internet access activity per day.

## **Inferential Analysis: Correlation & Regression**

#### **Correlation Analysis - The Society Literacy Development Index and The Level of Reading Interest**

Pearson correlation is a correlation measure used to measure the strength and direction of a linear relationship between two variables. Two variables are said to be correlated if a change in one variable is accompanied by a change in the other variable, either in the same direction or in the opposite direction. For this correlation analysis, our hypotheses are as follows.

H0: r = 0 (The correlation between the society literacy development index and the level of reading interest is not significant.)

H1: r ≠ 0 (The correlation between the society literacy development index and the level of reading interest is significant.)

α = 0.05

Test criteria: reject the null hypothesis if p-value < alpha

```python
from scipy.stats.stats import pearsonr

pearsonr(data['Society Literacy Development Index'], data2['Level of Reading Interest'])
```

From the Pearson's correlation analysis output, we get the Pearson correlation coefficient of 0,071 and the p-value for testing non-correlation of 0,687. The result shows that the p-value = 0,687 > α = 0,05. Therefore, we can not reject the null hypothesis, so we can conclude that **the correlation between the society literacy development index and the level of reading interest is not significant**.

#### **Regression Analysis - The Society Literacy Index and its Composer Elements**

Regression analysis generates an equation to describe the **statistical relationship** between one or more predictor variables and the response variable. Generally, linear regression is used **to find the effect** of one predictor variable on one response variable (simple linear regression) or the effect of several predictor variables on one response variable (multiple linear regression). In this case, we use the **multiple linear regression** with our response variable being the Society Literacy Development Index (Y), and our predictor variables are:
* Library service equity (X1)
* Adequacy of Library Collection (X2)
* Adequacy of Library Staff (X3)
* Society Visit Level per Day (X4)
* SNP-based Library (X5)
* Society Involvement in Socialization Activities (X6)
* Library Members (X7)

##### Variables Definition

```python
# Y variable
y = data['Society Literacy Development Index']
y.head()
```

```python
# X variables
X = data.drop(columns=['Province', 'Region', 'Society Literacy Development Index'], axis=1)
X.head()
```

##### Model Regression

Multiple linear regression analysis requires two tests: F-test (simulant) and t-test (individual).

* **F-test** in regression, often called the **Goodness of Fit test**, compares the fits of different linear models. Unlike t-tests that can assess only one regression coefficient at a time, the F-test can assess multiple coefficients simultaneously. Therefore, in the F-test of overall significance, the hypotheses are as follows.
    - H0: β1 = β2 = ... = β7 = 0 (all predictor variables simultaneously have no significant influence on the response variable)
    - H1: at least one βi ≠ 0 (there is at least one predictor variable that has a significant influence on the response variable)
      - α = 0.05
      - Test criteria: reject the null hypothesis if F-value > F-table or if p-value < alpha
* **t-test** in regression, known as the **partial test**, is used to test the influence of each predictor variable individually on the response variable. Therefore, in the t-test of individual significance, the hypotheses are as follows.
    - H0: βi = 0 (there is no significant influence between the response variable (Y) and one of the predictor variables (Xi))
    - H1: βi ≠ 0 (there is a significant influence between the response variable (Y) and one of the predictor variables (Xi))
      - α = 0.05
      - Test criteria: reject the null hypothesis if t-value > t-table or if p-value < alpha

```python
import numpy as np
import statsmodels.api as sm

X = sm.add_constant(X) # adding a constant
 
model = sm.OLS(y, X).fit()
predictions = model.predict(X) 
 
summary = model.summary()
print(summary)
```

The p-value for each term tests the null hypothesis that the coefficient equals zero (no effect). A low p-value (< 0.05) indicates that we can reject the null hypothesis. In other words, a predictor that has a low p-value is likely to be a meaningful addition (significant) to our model because changes in the predictor's value are related to changes in the response variable. Conversely, a larger (insignificant) p-value suggests that changes in the predictor are not associated with changes in the response.

The multiple linear regression analysis outputs show that all the p-values are less than alpha = 0,05 except for the intercept. This indicates that **all the predictor variables significantly influence the response variable**. In other words, Library service equity (X1), Adequacy of Library Collection (X2), Adequacy of Library Staff (X3), Society Visit Level per Day (X4), SNP-based Library (X5), Society Involvement in Socialization Activities (X6), and Library Members (X7), all have a significant influence on the Society Literacy Development Index (Y).


As a result, the multiple linear regression model is as follows.

Y = 85.8699 (X1) + 100.0355 (X2) + 164.5524 (X3) + 101.1480 (X4) + 86.7256 (X5) + 100.0473 (X6) + 99.9960 (X7)

Where:
* Y = Society Literacy Development Index
* X1 = Library Service Equity
* X2 = Adequacy of Library Collection
* X3 = Adequacy of Library Staff
* X4 = Society Visit Level per Day 
* X5 = SNP-based Library
* X6 = Society Involvement in Socialization Activities
* X7 = Library Members

For a continuous predictor variable, **the regression coefficient represents the difference in the predicted value of the response variable for each one-unit change in the predictor variable, assuming all other predictor variables are held constant**. Therefore, this model can be interpreted as follows.
- On average, each additional Library Service Equity (X1) is associated with an increase of 85.8699 points on the Society Literacy Development Index (Y), assuming all the other predictor variables are constant.
